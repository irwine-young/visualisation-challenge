### Analysis

1. There is a positive linear relationship (r=0.84) between average tumor volume and average mice weight.

2.  For capomulin treatment of mouse ID r554, there is significant reduction of tumor volume over treatment time (max treatment time at 45mins).

3.  Gender ratio for the mice in this study is almost even spread, 51:49 (male:female).

4.  On average, capomulin and ramicane's treatment (at max 45 minutes) reduces the tumor volume more than infubinol and ceftamin's treatment.

5.  Based on the boxplot, all 4s tumor volume follows a normal distribution, as such either mean or median tumor volume can be used to summarise the final tumor volume.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy import stats

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
# Merge left to larger data set (study_results)
mouse_data_complete = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID"])

# Display the data table for preview
mouse_data_complete.head(5)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

# Getting mean values, grouped by Drug Regimen and store as variable
mean_df = no_g989_df.groupby(["Drug Regimen"]).mean(numeric_only=True)
mean = mean_df["Tumor Volume (mm3)"]

# Getting median values, grouped by Drug Regimen and store as variable
median_df = no_g989_df.groupby(["Drug Regimen"]).median(numeric_only=True)
median = median_df["Tumor Volume (mm3)"]

# Getting variance, grouped by Drug Regimen and store as variable
variance_df = no_g989_df.groupby(["Drug Regimen"]).var(numeric_only=True)
variance = variance_df["Tumor Volume (mm3)"]

# Getting standard deviation, grouped by Drug Regimen and store as variable
stdev_df = no_g989_df.groupby(["Drug Regimen"]).std(numeric_only=True)
stdev = stdev_df["Tumor Volume (mm3)"]

# Getting SEM, grouped by Drug Regimen and store as variable
sem_df = no_g989_df.groupby(["Drug Regimen"]).sem(numeric_only=True)
sem = sem_df["Tumor Volume (mm3)"]

# Create dataframe for mean, median, variance, std dev and SEM
summary_statistics = mean_df[["Tumor Volume (mm3)"]]

# Rename column
summary_statistics = summary_statistics.rename(columns={
                                        "Tumor Volume (mm3)":"Mean Tumor Volume"})

# Append all other statistics values
summary_statistics["Median Tumor Volume"] = median
summary_statistics["Tumor Volume Variance"] = variance
summary_statistics["Tumor Volume Std. Dev."] = stdev
summary_statistics["Tumor Volume Std. Err."] = sem

# Print summary statistics
summary_statistics
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.

summary_statistics2 = no_g989_df.groupby(['Drug Regimen']).agg({'Tumor Volume (mm3)':
                                                                ['mean', 
                                                                 'median', 
                                                                 'var', 
                                                                 'std', 
                                                                 'sem']})
# Print summary statistics
summary_statistics2
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = mouse_data_complete[mouse_data_complete["Timepoint"] == 45]

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
last_timepoint_4drugs = last_timepoint[(last_timepoint["Drug Regimen"] == "Capomulin") | 
                                       (last_timepoint["Drug Regimen"] == "Ramicane") |
                                       (last_timepoint["Drug Regimen"] == "Infubinol") |
                                       (last_timepoint["Drug Regimen"] == "Ceftamin")]

last_timepoint_4drugs = last_timepoint_4drugs.rename(columns={'Timepoint': 'Max Timepoint'})
last_timepoint_4drugs.head(50)
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_volume = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in drugs:
   
# Locate the rows which contain mice on each drug and get the tumor volumes
    df = last_timepoint_4drugs.loc[last_timepoint_4drugs['Drug Regimen'] == drug]                             
    
    # add subset
    df2 = df.loc[df['Max Timepoint'] == 45]  
    
    # Determine Tumor Volume
    tumor_value = df2['Tumor Volume (mm3)']
    tumor_volume.append(tumor_value)   
    # Print to check tumor volume values
    print(tumor_volume)
    
    # Determine outliers using upper and lower bounds
    quartiles = tumor_value.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    print(f'IQR for {drug}: {iqr}')
    
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f'Lower Bound for {drug}: {lower_bound}')
    print(f'Upper Bound for {drug}: {upper_bound}')
    
    outliers_count = (tumor_value.loc[(df2['Tumor Volume (mm3)'] >= upper_bound) | 
                                        (df2['Tumor Volume (mm3)'] <= lower_bound)]).count()
    print(f'{drug} outliers: {outliers_count}')  
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig1, ax1 = plt.subplots()
ax1.set_title('Distribution of Tumor Volume vs Drug')
ax1.set_xlabel('Drug Regimen')
ax1.set_ylabel('Tumor Volume (mm3)')
ax1.set_xticklabels(['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])
ax1.boxplot(tumor_volume)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
# Get all data for Capomulin drug regimen
capomulin = no_g989_df[no_g989_df["Drug Regimen"] == "Capomulin"]

# The get all data for specific mouse ID r554
r554 = capomulin[capomulin["Mouse ID"] == "r554"]

x_axis = r554["Timepoint"]
y_axis = r554["Tumor Volume (mm3)"]

# Create a line chart based upon the above data
plt.plot(x_axis, y_axis)
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Tumor volume over time for Mouse ID r554 for Capomulin treatment")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

# Get all data for Capomulin drug regimen
capomulin = no_g989_df.loc[no_g989_df["Drug Regimen"] == "Capomulin"]

# Get mean tumor volume grouped by Mouse ID and convert to dataFrame
capomulin_tumor= capomulin.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
capomulin_tumor = pd.DataFrame(capomulin_tumor)

# Get mean weight grouped by Mouse ID and convert to dataFrame
capomulin_weight = capomulin.groupby('Mouse ID')['Weight (g)'].mean()
capomulin_weight = pd.DataFrame(capomulin_weight)

# Merge data into single dataFrame
capomulin_scatter = pd.merge(capomulin_tumor, capomulin_weight, how="left", on=["Mouse ID"])

x_scatter = capomulin_scatter["Weight (g)"]
y_scatter = capomulin_scatter["Tumor Volume (mm3)"]

# Create a scatter plot based upon the above data
plt.scatter(x_scatter, y_scatter)
plt.xlabel("Average Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Tumor volume vs mouse weight for Capomulin regimen")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

# Perform a linear regression 
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_scatter, y_scatter)

# Get regression values
regress_values = x_scatter * slope + intercept

# Create line equation string
line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))

# Create a scatter plot based upon the above data, include regression line and equation
plt.scatter(x_scatter, y_scatter)
plt.plot(x_scatter,regress_values,"r-")
plt.annotate(line_eq,(20,35),fontsize=15,color="red")
plt.xlabel("Average Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Tumor volume vs mouse weight for Capomulin regimen")
plt.show()

# Print r value
print(f"The r-value is: {rvalue}")
```
